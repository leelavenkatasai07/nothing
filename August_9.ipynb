```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.patches import Wedge
from matplotlib.collections import PatchCollection

labels=['C','C++','JAVA','PYTHON','DATASCIENCE']
prices=[400,450,1100,750,1350]
colors=['blue','grey','brown','magenta','green']

total=sum(prices)
fracs=[p/total for p in prices]
angles=[frac * 360 for frac in fracs]

fig=plt.figure(figsize=(8,8))
ax=fig.add_subplot(111,projection='3d')

radius=1
height=0.3
start_angle=0

patches = []
for i, angle in enumerate(angles):
    theta = np.linspace(np.deg2rad(start_angle), np.deg2rad(start_angle + angle), 30)
    x = np.append([0], radius * np.cos(theta))
    y = np.append([0], radius * np.sin(theta))
    ax.plot_trisurf(x, y, 0, color=colors[i], alpha=0.8, shade=False)
    ax.plot_trisurf(x, y, height, color=colors[i], alpha=0.8, shade=False)
    for j in range(len(theta) - 1):
        xs = [radius * np.cos(theta[j]), radius * np.cos(theta[j+1]), radius * np.cos(theta[j+1]), radius * np.cos(theta[j])]
        ys = [radius * np.sin(theta[j]), radius * np.sin(theta[j+1]), radius * np.sin(theta[j+1]), radius * np.sin(theta[j])]
        zs = [0, 0, height, height]
        verts = [list(zip(xs, ys, zs))]
        ax.add_collection3d(PatchCollection([plt.Polygon(v[:2]) for v in verts], color=colors[i], alpha=0.8), zs=zs, zdir='z')

mid_angle_rad = np.deg2rad(start_angle + angle/2)
ax.text(radius * 0.7 * np.cos(mid_angle_rad), radius * 0.7 * np.sin(mid_angle_rad), height + 0.05, f"{labels[i]}:Rs.{prices[i]}", ha='center', va='bottom', fontsize=10, color='black')
start_angle += angle
ax.set_box_aspect([1,1,0.5])
ax.axis('off')
plt.title("3D Pie Chart of Book Prices")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

hours = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
marks = np.array([2, 4, 6, 4, 5])

model = LinearRegression()
model.fit(hours, marks)

new_hours = np.array([6]).reshape(-1, 1)
prediction = model.predict(new_hours)

print(f"Predicted marks for 6 hours: {prediction[0]:.2f}")

plt.scatter(hours, marks, color='blue', label='Original Data')
plt.plot(hours, model.predict(hours), color='red', label='Regression Line')
plt.xlabel('Hours')
plt.ylabel('Marks')
plt.title('Hours vs Marks')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
np.random.seed(42)
x=np.random.rand(100)*10
y=3*x +5+np.random.randn(100)*2
m,c=np.polyfit(x,y,1)
print(f"slope(m):{m:.2f}")
print(f"intercept(c):{c:.2f}")
newx=7
print(f"Predicted y for x={newx}:{m*newx+c:.2f}")
#plt.title("Linear Regression")
plt.scatter(x,y,color='blue',alpha=0.6,label='Data Point')
plt.plot(x,m*x+c,color='red',lw=2,label='Best Fitting Line')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
x=np.array([[5,2],[3,1],[8,3],[2,0],[6,2]])
y=np.array([72,50,90,30,85])
model=LinearRegression()
model.fit(x,y)
print("Intercept:",model.intercept_)
print("Coeffcecent:",model.coef_)
y_pred=model.predict(x)
plt.scatter(y,y_pred,color='blue')
plt.plot([y.min(),y.max()],[y.min(),y.max()],'r--',lw=2)
plt.xlabel('Actual Marks')
plt.ylabel('Predicted Marks')
plt.title('original vs predited')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from mpl_toolkits.mplot3d import Axes3D
np.random.seed(42)
num_points = 800
theta=np.linspace(0,6*np.pi,num_points)
r=np.linspace(0,50,num_points)+np.random.normal(0,2,num_points)
z=np.linspace(-10,10,num_points)
x=r*np.cos(theta)
y=r*np.sin(theta)
r_value,p_value=pearsonr(x,y)
fig=plt.figure(figsize=(8,8))
ax=fig.add_subplot(111,projection='3d')
sc=ax.scatter(x,y,z,c=theta,cmap='plasma',s=5,alpha=0.6)
#plt.figure(figsize=(7,7))
plt.scatter(x,y,c=theta,cmap='plasma',s=5,alpha=0.6)
plt.title(f"Pearson plot={r_value:.3f}",fontsize=15)
plt.axis("off")
plt.colorbar(label='angle Theta')
plt.show()
```
