```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
np.random.seed(42)
X1=np.random.rand(50)*100
Y1=np.random.rand(50)*100
Z1=np.random.rand(50)*100
colors1=np.random.rand(50)
sizes1=np.random.rand(50)*500
X2=np.random.rand(40)*100
Y2=np.random.rand(40)*100
Z2=np.random.rand(40)*100
colors2=np.random.rand(40)
sizes2=np.random.rand(40)*500
fig=plt.figure(figsize=(12,8))
ax=fig.add_subplot(111,projection='3d')
scatter1=ax.scatter(X1,Y1,Z1,c=colors1,s=sizes1,cmap='viridis',edgecolor='black',marker='o',alpha=0.7,label='dataset1')
scatter2=ax.scatter(X2,Y2,Z2,c=colors2,s=sizes2,cmap='plasma',edgecolor='black',marker='^',alpha=0.7,label='dataset2')
cbar1=fig.colorbar(scatter1,ax=ax,shrink=0.5, pad=0.1)
cbar1.set_label("Viridis")
cbar2=fig.colorbar(scatter2,ax=ax,shrink=0.5, pad=0.09)
cbar2.set_label("Plasma")
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_zlabel('z-axis')
plt.title('3D Random Scatter Plot with 2 datasets')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
np.random.seed(42)
data1=np.random.randint(1,100,50)
data2=np.random.randint(1,100,50)
data3=np.random.randint(1,100,50)
bins=10
count1,bin_edges=np.histogram(data1,bins=bins,range=(0,100))
count2,bin_edges=np.histogram(data2,bins=bins,range=(0,100))
count3,bin_edges=np.histogram(data3,bins=bins,range=(0,100))
bin_centers=(bin_edges[:-1]+bin_edges[1:])/2
fig=plt.figure(figsize=(12,8))
ax=fig.add_subplot(111,projection='3d')
dx=dy=(bin_edges[1]-bin_edges[0])
x1=bin_centers-dx
x2=bin_centers
x3=bin_centers+dx
y1=np.zeros_like(x1)
y2=np.zeros_like(x2)*1.5
y3=np.zeros_like(x3)*3
ax.bar3d(x1,y1,np.zeros_like(count1),dx,dy,count1,color='blue',alpha=0.8,label='Dataset1')
ax.bar3d(x2,y2,np.zeros_like(count2),dx,dy,count2,color='green',alpha=0.8,label='Dataset2')
ax.bar3d(x3,y3,np.zeros_like(count3),dx,dy,count3,color='black',alpha=0.8,label='Dataset3')
ax.set_xlabel("Values")
ax.set_ylabel("Frequency")
ax.set_title("multiple Histogram overlapping")
ax.set_yticks([0,1.5,3])
ax.set_yticklabels(['Dataset1','Dataset2','Dataset3'])
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
labels=['C','C++','JAVA','PYTHON','DATASCIENCE']
prices=[400,450,1100,750,1350]
colors=['blue','grey','brown','magenta','green']
explode=[0,0,0.2,0,0.2]
def label_prices(pct,all_vals):
  absolute=int(round(pct/100*sum(all_vals)))
  return f'Rs.{absolute}'
plt.figure(figsize=(8,8))
wedges,texts,autotexts=plt.pie(prices,labels=labels,colors=colors,explode=explode,autopct=lambda pct:label_prices(pct,prices),pctdistance=0.6,startangle=90,textprops={'fontsize':12,'fontweight':'bold','color':'black'})
plt.title("books prices",fontsize=16,fontweight='bold')
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.patches import Wedge
from matplotlib.collections import PatchCollection

labels=['C','C++','JAVA','PYTHON','DATASCIENCE']
prices=[400,450,1100,750,1350]
colors=['blue','grey','brown','magenta','green']

total=sum(prices)
fracs=[p/total for p in prices]
angles=[frac * 360 for frac in fracs]

fig=plt.figure(figsize=(8,8))
ax=fig.add_subplot(111,projection='3d')

radius=1
height=0.3
start_angle=0

patches = []
for i, angle in enumerate(angles):
    theta = np.linspace(np.deg2rad(start_angle), np.deg2rad(start_angle + angle), 30)
    x = np.append([0], radius * np.cos(theta))
    y = np.append([0], radius * np.sin(theta))
    ax.plot_trisurf(x, y, 0, color=colors[i], alpha=0.8, shade=False)
    ax.plot_trisurf(x, y, height, color=colors[i], alpha=0.8, shade=False)
    for j in range(len(theta) - 1):
        xs = [radius * np.cos(theta[j]), radius * np.cos(theta[j+1]), radius * np.cos(theta[j+1]), radius * np.cos(theta[j])]
        ys = [radius * np.sin(theta[j]), radius * np.sin(theta[j+1]), radius * np.sin(theta[j+1]), radius * np.sin(theta[j])]
        zs = [0, 0, height, height]
        verts = [list(zip(xs, ys, zs))]
        ax.add_collection3d(PatchCollection([plt.Polygon(v[:2]) for v in verts], color=colors[i], alpha=0.8), zs=zs, zdir='z')

mid_angle_rad = np.deg2rad(start_angle + angle/2)
ax.text(radius * 0.7 * np.cos(mid_angle_rad), radius * 0.7 * np.sin(mid_angle_rad), height + 0.05, f"{labels[i]}:Rs.{prices[i]}", ha='center', va='bottom', fontsize=10, color='black')
start_angle += angle
ax.set_box_aspect([1,1,0.5])
ax.axis('off')
plt.title("3D Pie Chart of Book Prices")
plt.show()
```
